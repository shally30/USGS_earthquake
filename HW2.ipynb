```python
import json
import pandas as pd
import datetime
```

```python
json_data = json.load(open('/Users/zhangling/Desktop/earthquake-all.json'))
json_data
```

```python
json_lean = []
for f in json_data['features']:
    json_lean.append({
        'longitude': f['geometry']['coordinates'][0],
        'latitude': f['geometry']['coordinates'][1],
        'elevation': f['geometry']['coordinates'][2],
        'magnitude': f['properties']['mag'],
        'magnitude^10': (f['properties']['mag'])**10,
        'place': f['properties']['place'],
        'id': f['id'],
        'time': f['properties']['time'],
    })
```

```python
df = pd.DataFrame(json_lean)
df.head()
```

```python
json.dump(json_lean, open('/Users/zhangling/Desktop/7910 Tableau/Homework/HW2-earthquake/earthquake-lean.json', 'w'))
```

```python
df['dt'] = df['time'].apply(
    lambda t: datetime.datetime.fromtimestamp(t/ 1000)
)
```

```python
df['year'] = df['dt'].apply(lambda dt: dt.year)
```

```python
df.to_csv('earthquake-lean.csv', index=None)
```
